```python
from transformers import GPT2LMHeadModel
```

```python
model_hf = GPT2LMHeadModel.from_pretrained("gpt2")
print("Model loaded successfully:", model_hf)
```

```python
# Print model state dict keys and their shapes - dictionary, these are the raw tensors from the model
sd_hf = model_hf.state_dict()

# find max key length and print keys and shapes aligned
max_key_length = max(len(k) for k in sd_hf.keys())
for k, v in sd_hf.items():
    print(f"{k:<{max_key_length}} {v.shape}")
```
